```r
envardf<-outdf[,c("id", "lat","lon")]
for (i in 1:5){
    meanr <- runif(1, -10, 100)
    sdrex <- runif(1, -3, 3)
    sdr<-10^sdrex
    print(sdr)
    varname <- paste0("envar_", i)
    values <- rnorm(
        nrow(envardf),
        mean = meanr,
        sd = sdr
    )
    envardf[,varname] <- values
}
```

```r
save_test_path<-function(x = NULL){
    if (is.null(x)) {
        outvect <- paste(getwd(), "test-data",sep="/")
        return(outvect)
    }
    return(paste(getwd(), "test-data", x,sep="/"))
    
}
```
